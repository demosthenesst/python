# Θεώρημα ακεραίων ριζών πολυωνύμου


```python
from IPython.display import display, Math, clear_output
from sympy import *
from sympy.abc import x
import random




integerRootPols = [x-r for r in range(-6,7)]
fractionalRootPols = [2*x-1, 2*x-3, 2*x-5,
                           3*x-1, 3*x-2, 3*x-4, 3*x-5,
                           4*x-1, 4*x-3, 4*x-5,
                           5*x-1, 5*x-2, 5*x-3, 5*x-4,
                           2*x+1, 2*x+3, 2*x+5,
                           3*x+1, 3*x+2, 3*x+4, 3*x+5,
                           4*x+1, 4*x+3, 4*x+5,
                           5*x+1, 5*x+2, 5*x+3, 5*x+4
                    ]

noRootPols = [expand( (x-a)**2 +b ) for a in range(-4, 4) for b in range(1,3)]
#noRootPols = [x**2+1, x**2+2, x**2+3, x**2+x+1, x**2 -x+1, x**2+2*x+3, x**2-2*x+3  ]




numOfintRoots = random.randint(2,3) 
numOfFractionalRoots = random.randint(0,1)
numOfnoRoots = 1-numOfFractionalRoots # number of irreducibles 

fracPols = random.choices(fractionalRootPols, k=numOfFractionalRoots)
irrPols = random.choices(noRootPols, k=numOfnoRoots)
intPols = random.choices(integerRootPols, k=numOfintRoots)

polList = fracPols + irrPols + intPols




factoredPol = prod([p for p in polList])
constCoeff = factoredPol.subs(x,0)
polRootsSet = solveset(factoredPol,x, domain=S.Reals)

if abs(constCoeff)>90:
    m = abs( min(polRootsSet) )
    M = max(polRootsSet)
    if m<M:
        factoredPol = factoredPol/(x-M)
    else:
        factoredPol = factoredPol/(x+m)

pol = expand(factoredPol)
polRootsSet = solveset(factoredPol,x, domain=S.Reals)
POLSTRING = latex(pol)
FPOLSTRING = latex(factoredPol)



#display(polRootsSet)
#MM=max( max(polRootsSet), abs( min(polRootsSet) ) )
#display(MM)



while len(solveset(pol,x, domain=S.Reals))>0:
    display( Math('P(x)=%s' %latex(pol) ) )
    print('Εισαγωγή ρίζας:')
    rinput = input('')
    rinput = sympify(rinput)
    clear_output()
    if rinput in solveset(pol,x, domain=S.Reals):
        print('\nΣωστά!\n')
        pol = simplify(pol/(x-rinput))
        #display( Math('P(x)=\\left(%s\\right)\\cdot\\left(%s\\right)' %( latex(pol), latex(x-rinput) ) ) )
        #display( Math('P(x)=%s' %latex(pol) ) )
    else:
        print('\nΞαναπροσπάθησε.')  


if numOfnoRoots==1:
    while True:
        display(irrPols[0])
        print('Εισαγωγή ρίζας:')
        inp = input('')
        if inp == ' ':
            print('Σωστά')
            break
        else:
            print('\nΞαναπροσπάθησε.')
        

clear_output()

print('\nΤο σύνολο των ριζών του πολυωνύμου είναι:')
display(polRootsSet)
print('\nΗ παραγοντοποίηση του αρχικού πολυωνύμου είναι:')
display( Math('%s=%s' %(POLSTRING, FPOLSTRING )  ) )


```
