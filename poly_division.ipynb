# Εξάσκηση στη διαίρεση πολυωνύμων

Στην παρακάτω εφαρμογή ζητείται η εισαγωγή δύο πολυωνύμων (του διαιρετέου $\Delta(x)$ και του διαιρέτη $\delta(x)$) και εμφανίζονται (σχεδόν πλήρως) τα βήματα του αλγορίθμου της διαίρεσής τους. Εμφανίζονται επίσης το πηλίκο $\pi(x)$ και το υπόλοιπο $\upsilon(x)$.

- Η εισαγωγή των πολυωνύμων γίνεται δεκτή από τους συντελεστές τους. Παραδείγματος χάριν "-2,5,sqrt(3),1/4, 2.7" για το πολυώνυμο $-2x^4 +5x^3 +\sqrt{3}x^2 +\frac{1}{4}x+2,7$.
- Εισάγοντας έναν κενό χαρακτήρα, η εφαρμογή τερματίζεται 

```python

from IPython.display import display, Math, clear_output
from sympy import *
#from sympy import poly
from sympy.abc import x
import numpy

def polynomialDivisionData(D,d):
    divident = D
    #list = [[D,d]]
    list=[]
    quotient = 0
    if degree(D,gen=x)<degree(d,gen = x):
        list = [[D,0]]
    else:
        while degree(divident,gen=x)>=degree(d,gen = x):
            lt = LT(poly(divident,x))/LT(poly(d,x))
            quotient = quotient+lt
            divident = divident - d*lt
            divident = expand(divident)
            list.append([divident,quotient])
    return list

def matToArr(st): ### latex replacements matrix to array
    st=st.replace('\\text{None}', '\\quad')
    st=st.replace('\\left[\\begin{matrix}', '\\begin{array}{r|l}').replace('\\end{matrix}\\right]', '\\end{array}')
    #st = '\\\\ \\hline '.join(st.rsplit('\\\\', 2))
    #st = '\\\\'.join(st.rsplit('\\\\ \\hline ', 1))
    st = st.replace("\\\\", "\\\\ \\hline", 1)
    return st

print('\nΕισάγετε τους συντελεστές του πολυωνύμου-διαιρετέου:')

termination = False

while True:
    inp1 = input('')
    if inp1 == ' ':
        termination = True
        clear_output()
        break
    while 'x' in inp1:
        print('\nΗ εισαγωγή δεν πρέπει να περιέχει "x".')
        inp1 = input('')
    else:
        coeffInput1 = inp1.replace(" ","").split(",")
    try: 
        sympify(coeffInput1)
    except ValueError:
        print('\nΗ εισαγωγή συντελεστών δεν είναι έγκυρη. Δοκιμάστε ξανά.')
    else:
        break

clear_output()

if termination == False:
    print('\nΕισάγετε τους συντελεστές του πολυωνύμου-διαιρέτη:')

while termination == False:
    inp2 = input('')
    if inp2 == ' ':
        termination = True
        clear_output()
        break
    while 'x' in inp2:
        print('\nΗ εισαγωγή δεν πρέπει να περιέχει "x".')
        inp2 = input('')
    else:
        coeffInput2 = inp2.replace(" ","").split(",")
    try: 
        sympify(coeffInput2)
    except ValueError:
        print('\nΗ εισαγωγή συντελεστών δεν είναι έγκυρη. Δοκιμάστε ξανά.')
    else:
        if len(coeffInput2)==1 and coeffInput2[0]=='0':
            print('\nO διαιρέτης δεν πρέπει να είναι το μηδενικό πολυώνυμο.')
        else:
            break

while termination == False:
    coeffList1 = sympify(coeffInput1)
    coeffList2 = sympify(coeffInput2)
    
    DIVIDENT = Poly.from_list(coeffList1, gens=x).as_expr()
    DIVISOR = Poly.from_list(coeffList2, gens=x).as_expr()
    
    divisionData = polynomialDivisionData(DIVIDENT,DIVISOR)
    noOfSteps = len(divisionData)
    
    divisionArray = numpy.empty((noOfSteps+1,2), dtype=object)
    
    divisionArray[0][0] = DIVIDENT
    divisionArray[0][1] = DIVISOR
    
    clear_output()
    print('\n\u2022 Διαιρετέος:')
    display(DIVIDENT)
    print('\n\u2022 Διαιρέτης:')
    display(DIVISOR)
    
    print('\n\nΑκολουθούν τα βήματα εκτέλεσης του αλγορίθμου της διαίρεσης.')
    for i in range(0, noOfSteps):
        divisionArray[i+1][0]= divisionData[i][0]
        divisionArray[1][1]= divisionData[i][1]
        ltx = latex(sympify(divisionArray))
        print('\n\u2022 Βήμα '+str(i+1)+'o:')
        display(Math(matToArr(ltx)))
    
    [REMAINDER, QUOTIENT] = divisionData[-1]
    
    print('\n\u2022 Πηλίκο:')
    display(QUOTIENT)
    print('\n\u2022 Υπόλοιπο:')
    display(REMAINDER)
    print('\n\u2022 Ταυτότητα διαίρεσης:')
    
    remlx = latex(REMAINDER)
    if latex(REMAINDER)[0]=='-':
        remlx = '\\left('+latex(REMAINDER)+'\\right)'
    
    divsrlx = latex(DIVISOR)
    if len(latex(DIVISOR))>1:
        divsrlx = '\\left('+latex(DIVISOR)+'\\right)'
    
    display(  Math(  '%s = %s\\cdot\\left(%s\\right)+%s' %( latex(DIVIDENT), divsrlx, latex(QUOTIENT), remlx )  )  )
    break
```
