
# Εξάσκηση στο σχήμα Horner

Στόχος της ακόλουθης δραστηριότητας είναι η εξάσκηση στο σχήμα Horner. Με την εκτέλεση της εφαρμογής ζητείται κάθε φορά η σωστή εισαγωγή στην αντίστοιχη θέση του σχήματος Horner και εμφανίζεται το ανάλογο μήνυμα αν είναι σωστή ή όχι η εκάστοτε επιλογή.

Ανα πάσα στιγμή:

- Η εφαρμογή τερματίζεται πληκτρολογώντας "Χ" στο πεδίο εισαγωγής.
- Η τρέχουσα προσπάθεια εγκαταλείπεται πληκτρολογώντας "Α" στο πεδίο εισαγωγής. Παρουσιάζεται το πλήρες σχήμα και δίνεται η δυνατότητα για νέα προσπάθεια.

```python

from IPython.display import display, Math, clear_output
#from sympy import symbols, latex, factor, simplify, sympify
import random
from sympy import *
import numpy
from time import sleep

def hornerAns (coeffs, val):
    p = coeffs[0]
    q = 0
    i = 1
    list = []
    while i<len(coeffs):
        q = p * val
        p = q + coeffs[i]
        list.append(q)
        list.append(p)
        i=i+1
    return list

def matToArr(st, col): ### latex replacements matrix to array
    c=col*'r'
    st=st.replace('\\text{None}', '\\quad')
    st=st.replace('\\left[\\begin{matrix}', '\\begin{array}{'+c+'|r}').replace('\\end{matrix}\\right]', '\\end{array}')
    st = '\\\\ \hline '.join(st.rsplit('\\\\', 1))  #replaces last occurence
    st = ''.join(st.rsplit('\\quad\\quad\\quad', 2))
    return st

###############################################

termination = False

while True:

    while True: #keeping values small
        degree = random.randint(2,6)
        lc = random.randint(1,6)
        coeffList = [lc] + [random.randint(-8,8) for i in range(0,degree)]
        value = random.randint(-6,6)
        
        if value == 0:
            value = 1
        
        if hornerAns(coeffList, value)[-1] <= 150 and hornerAns(coeffList, value)[-1] >= -150:
            break

    hornerAnswers = hornerAns(coeffList, value)
    answersLength = len(hornerAnswers)
    
    hornerArray = numpy.empty((3,degree+2), dtype=object)
    hornerArray[0,-1] = value
    hornerArray[2,0]= coeffList[0]
    
    for i in range(0,degree+1):
        hornerArray[0,i]=coeffList[i]
    
    printStrings = [latex(sympify(hornerArray))]
    
    for i in range(0, answersLength):
        if i%2 == 0:
            hornerArray[1,int((i+2)/2)] = hornerAnswers[i]
            printStrings.append(latex(sympify(hornerArray)))
        if i%2 == 1:
            hornerArray[2,int((i+1)/2)] = hornerAnswers[i]
            printStrings.append(latex(sympify(hornerArray)))
    
    latexStrings = [matToArr(x, degree+1) for x in printStrings]

    clear_output()

    step=0

    while step < answersLength:
        ltx = latexStrings[step]
        print()
        display(Math(ltx))
        inp = input('\nΕπόμενη καταχώρηση:\n')
        if inp == str(hornerAnswers[step]):
            #clear_output()
            print('\n\u2714 Σωστά!')
            sleep(2)
            clear_output()
            step = step+1
        elif inp in ['X', 'x', 'Χ', 'χ']:
            clear_output()
            print('\nΤερματισμός εφαρμογής.')
            termination = True
            break
        elif inp in ['A', 'a', 'Α', 'α']:
            sleep(1)
            clear_output()
            print('\nΕγκατάλειψη προσπάθειας.')
            break
        else:
            print('\n\u2716 Ξαναπροσπάθησε.')
            sleep(2)
            clear_output()

    print('\nΤο τελικό σχήμα είναι:\n')
    display(Math(latexStrings[-1]))
    
    if termination == True:
        break
    
    if inp not in ['A', 'a', 'Α', 'α']:
        print('\nΣυγχαρητήρια για την επιτυχή προσπάθεια!')
    
    newGame = input('\n Επιθυμείς νέα προσπάθεια; (Ν/Ο)\n')
    if newGame not in ['N', 'n', 'Ν', 'ν']:
        clear_output()
        break
```
